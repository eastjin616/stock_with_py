```python
#dataframe만드는 방법
import pandas as pd

df = pd.DataFrame()
df['Name'] = ["Jacky","Steven","George"]
df['Age'] = [23, 34, 29]
df['driver'] = [True, False, True]

df
```

```python
data = {
    "Name" : ["Jacky","Steven","George"],
    "Age" : [23,34,29],
    "driver" : [True, False, True]
}
df = pd.DataFrame(data)
df.index = ["a","b","c"]
df.index.name = "index"

df
```

```python
#새로운 column data 추가하기
df["LOCATION"] = ["USA", "UK", "Canada"]
df
```

```python
#column 이름 변경하기
df = df.rename(columns={"Name":"Person"})
df
```

```python
#column 삭제하기
df.drop(columns="LOCATION", inplace=True)
df
```

```python
#새로운 row를 dataframe형태로 만들기
data = {
    "Person" : ["Harry"],
    "Age" : [25],
    "driver" : [False]
}

new_df = pd.DataFrame(data, index=["d"])
new_df
```

```python
#새로운 데이터를 dataframe형태로 만들기
df = pd.concat([df, new_df])
df
```

```python
data = {
    "Name" : ["Jacky","Steven","George","Harry"],
    "Age" : [23,34,29,105],
    "driver" : [True, False, True, False]
}
df = pd.DataFrame(data)
df.index = ["a","b","c","d"]
df.index.name = "index"

df
```

```python
#대괄호 인덱싱 [] 
    # - 특정 열을 선택할 때 사용 
    # - 슬라이싱을 사용하여 행을 선택할 수 있음 예 df[1:5]

#1 특정 열 선택하기
df['Name']
```

```python
#여러개의 열
df[['Name','Age']]
```

```python
#2 여러 행 데이터 선택
df[0:2]
  
```

```python
#.lock[]인덱서를 사용해서 index가 'a'인 행 접근하기
df.loc['a']
```

```python
#.loc을 사용하여 indexrk 'a'인 행에서 'Name'열에 접근하기
df.loc['a','Name']
```

```python
#.iloc[] 인덱서를 사용해서 위치기반 인덱싱 하기
    # - 행과 열의 위치(정수 인덱스)를 사용하여 데이터
df
```

```python
#,iloc 를 사용하여 두 번째 행에 접근하기
df.iloc[1]
```

```python
#.iloc을 사용하여 2번째 행, 1번째 열에 접근하기
print(df.iloc[1,1])
print(df.iloc[2,0])
```

```python
#불리언 인덱싱
#조건에 맞는 행 선택하기
# df[df['Age'] > 30]
df
```

```python
#4 boolean 인덱싱은 데이터를 필터링 할 때 많이 사용됨
df['Age'] > 25
```

```python
df[df['Age'] > 25]
```

```python
#.at[] ,.iat[] 인덱서를 사용하여 스칼라 값에 접근하기
    # - .at[] : 라벨 기반 인덱싱
    # - .iat[] : 위치 기반 인덱싱

df
```

```python
#at을 사용하여 index 'a'의 'Name'열에 접근하기
df.at['a','driver']
```

```python
#.iloc[] 인덱서를 사용해서 위치기반 인덱싱 하기
df.iat[0,2]
```
